```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
product_sales.sort_values(ascending=False).plot(kind='bar', figsize=(8,4))
plt.title("Revenue by Product")
plt.xlabel("product")
plt.ylabel("revenue")
plt.show()
```

```python
df['paymentmethod'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    figsize=(6,6)
)
plt.title("Payment Method Distribution")
plt.ylabel("")
plt.show()
```

```python
df['product'].value_counts().plot(kind='bar',figsize=(6,3))
plt.title("Total Product")
plt.xlabel("product")
plt.ylabel("total")
plt.show()
```

```python
monthly_sales = df.groupby('month')['totalprice'].sum()
monthly_sales.plot(kind='line', marker='o', figsize=(4,3))
plt.title("Monthly Sales Trend")
plt.xlabel("month")
plt.ylabel("sales")
plt.show()
```

```python
yearly_sales = df.groupby('year')['totalprice'].sum()
yearly_sales.plot(kind='bar', figsize=(4,3))
plt.title("Yearly Sales")
plt.xlabel("year")
plt.ylabel("sales")
plt.show()
```

```python
df['orderstatus'].value_counts().plot(kind='bar', figsize=(4,3))
plt.title("Order Status Count")
plt.xlabel("orderstatus")
plt.ylabel("count")
plt.show()
```

```python
df['paymentmethod'].value_counts().plot(kind='bar', figsize=(4,3))
plt.title("Payment Method Count")
plt.xlabel("paymentmethod")
plt.ylabel("count")
plt.show()
```
