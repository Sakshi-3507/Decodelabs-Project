# SQL Project Analysis

Professional notebook for SQL internship task.

## Create Database

```
%%sql
CREATE DATABASE ecommerce_db;
```

## Use Database

```
%%sql
USE ecommerce_db;
```

## Create Table

```
%%sql
CREATE TABLE orders (
    OrderID TEXT,
    Date TEXT,
    CustomerID TEXT,
    Product TEXT,
    Quantity TEXT,
    UnitPrice TEXT,
    ShippingAddress TEXT,
    PaymentMethod TEXT,
    OrderStatus TEXT,
    TrackingNumber TEXT,
    ItemsInCart TEXT,
    CouponCode TEXT,
    ReferralSource TEXT,
    TotalPrice TEXT
);
```

## SQL Query

```
%%sql
SELECT * FROM orders LIMIT 10;
```

## SQL Query

```
%%sql
SELECT COUNT(*) AS TotalOrders FROM orders;
```

## SQL Query

```
%%sql
SELECT SUM(TotalPrice) AS TotalRevenue FROM orders;
```

## SQL Query

```
%%sql
SELECT AVG(TotalPrice) AS AverageOrderValue FROM orders;
```

## SQL Query

```
%%sql
SELECT MAX(TotalPrice) AS HighestOrder FROM orders;
```

## SQL Query

```
%%sql
SELECT Product,SUM(TotalPrice) AS Revenue
FROM orders
GROUP BY Product
ORDER BY Revenue DESC;
```

## SQL Query

```
%%sql
SELECT PaymentMethod,COUNT(*) AS UsageCount
FROM orders
GROUP BY PaymentMethod
ORDER BY UsageCount DESC;
```

## SQL Query

```
%%sql
SELECT OrderStatus,COUNT(*) AS TotalOrders
FROM orders
GROUP BY OrderStatus
ORDER BY TotalOrders DESC
LIMIT 1;
```

## SQL Query

```
%%sql
SELECT MONTH(Date) AS Month,SUM(TotalPrice) AS MonthlySales
FROM orders
GROUP BY MONTH(Date)
ORDER BY Month;
```

## SQL Query

```
%%sql
SELECT ReferralSource,COUNT(*) AS Orders
FROM orders
GROUP BY ReferralSource
ORDER BY Orders DESC;
```

## SQL Query

```
%%sql
SELECT CustomerID,SUM(TotalPrice) AS TotalSpent
FROM orders
GROUP BY CustomerID
ORDER BY TotalSpent DESC
LIMIT 10;;
```
